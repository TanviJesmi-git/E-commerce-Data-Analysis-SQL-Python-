```python
decimal_cols = [
    'product_weight_g',
    'product_length_cm',
    'product_height_cm',
    'product_width_cm'
]

int_cols = [
    'product_name_lenght',
    'product_description_lenght',
    'product_photos_qty'
]

products[decimal_cols] = (
    products[decimal_cols]
    .astype(float)
    .fillna(0.0)
)

products[int_cols] = (
    products[int_cols]
    .astype('Int64')
    .fillna(0)
)
```

```python
products['product_category_name'] = (
    products['product_category_name']
    .fillna('unknown')
    .str.strip()
)
```

```python
orders=pd.read_csv("/content/olist_orders_dataset.csv")

# Identify string columns (object dtype)
object_cols = orders.select_dtypes(include='object').columns

print("Stripping whitespace and checking for empty strings in object columns:")
for col in object_cols:
    # Strip whitespace from the column. NaN values will remain NaN.
    orders.loc[:, col] = orders[col].str.strip()

    # Check for empty strings after stripping. NaN values are not counted as empty strings.
    empty_string_count = (orders[col] == '').sum()
    if empty_string_count > 0:
        print(f"Column '{col}': {empty_string_count} empty string(s) found after stripping.")
    else:
        print(f"Column '{col}': No empty strings found after stripping.")
```
